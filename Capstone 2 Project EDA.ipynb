```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_excel('C:/Users/Samuel Kim/Downloads/HospitalCosts.xlsx' )
sns.set(style='ticks')
```

```python
sns.stripplot(data=df,
              x='AGE')
plt.xlabel('Age')
plt.ylabel('Length of Stay')
plt.title('Length of Stay based on age')
sns.despine()
plt.show()
```

```python
sns.stripplot(data=df,
            x='AGE',
            y='LOS')
plt.xlabel('AGE')
plt.ylabel('LOS')
sns.despine()
plt.show()
```

```python
sns.stripplot(data=df,
             x='FEMALE',
             y='LOS')
plt.xticks(rotation=-15)
plt.xlabel('Gender')
plt.ylabel('Length of Stay')
sns.despine()

plt.show()
```

```python
#0 represent male 1 represnet female
sns.countplot(data=df,
              x='FEMALE')
plt.xlabel('Gender')
plt.ylabel('Length of Stay')
plt.title('Length of Stay based on age')
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,8),dpi=100)
sns.boxplot(data=df, x='AGE', y='TOTCHG')
```
